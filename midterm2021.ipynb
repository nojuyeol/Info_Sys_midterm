기말시험 면제자(기말고사 만점)는 과제1, 과제2, 중간고사 점수의 합계가 높은 순서로 3명이다. 동점인 경우, 중간고사 성적이 높은 사람이 우선권을 가지며, 그래도 동점인 경우, 중간고사 답안 조기 제출자가 우선권을 갖는다.

본 수업에서 배운 아래 방법을 사용해야 한다. 
- NN package: PyTorch (TensorFlow 안됨)
- loss: CrossEntropyLoss()
- optimizer: SGD, Adam
- activation function: Relu, Leaky Relu, Tanh, Sigmoid
- Layer: Linear, Conv2d, Maxpool2d, Avgpool2d, Dropout, BatchNorm1d
- momentum 사용불가
- xavier_uniform_ 사용 가능
- GPU 사용 가능

- Conv2d layer의 수는 2개 이하로 하며, Linear layer 수 2개 이하로 하여야 한다.
- 주어진 입력 이미지를 (64,64)로 resize 한 후에 사용하여야 한다.
- epoch는 20 이하
- learning rate는 제한 없음
- 주어진 training set을 나누어 training set과 validation set으로 나누어 사용할 수 있다.
